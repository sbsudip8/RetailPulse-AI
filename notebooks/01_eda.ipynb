```python
# Cell 1 â Imports & Load
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
df = pd.read_csv('../data/OnlineRetail.csv', encoding='ISO-8859-1')
df.head()
```

```python
# Cell 2 â Shape, dtypes, nulls
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

```python
# Cell 3 â Distribution plots
df['Quantity'].describe()
df['UnitPrice'].describe()
sns.histplot(df['UnitPrice'].clip(0, 50))
plt.show()
```

```python
# Cell 4 â Correlation heatmap (numeric columns only)
sns.heatmap(df[['Quantity','UnitPrice']].corr(), annot=True)
plt.show()
```

```python
# Cell 5 â Top countries by revenue
df['Revenue'] = df['Quantity'] * df['UnitPrice']
df.groupby('Country')['Revenue'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.show()
```
