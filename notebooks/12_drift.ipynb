```python

from evidently import Report
from evidently.presets import DataDriftPreset
import pandas as pd

rfm = pd.read_csv('../data/rfm_segmented.csv')

# Split into reference (first 70%) and current (last 30%) to simulate drift
ref     = rfm.iloc[:int(len(rfm)*0.7)][['Recency','Frequency','Monetary']]
current = rfm.iloc[int(len(rfm)*0.7):][['Recency','Frequency','Monetary']]

report = Report(metrics=[DataDriftPreset()])
snapshot = report.run(reference_data=ref, current_data=current)
snapshot.save_html('../data/drift_report.html')
print("Drift report saved ✓")
```
