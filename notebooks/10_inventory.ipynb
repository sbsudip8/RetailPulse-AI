```python
import pandas as pd, numpy as np

forecast = pd.read_csv('../data/prophet_forecast.csv', parse_dates=['ds'])
next_30 = forecast.tail(30)

# Simple reorder point logic
avg_daily_demand = next_30['yhat'].mean()
lead_time_days   = 3          # assume 3-day supplier lead time
safety_stock     = avg_daily_demand * 1.5   # 1.5x buffer

reorder_point  = (avg_daily_demand * lead_time_days) + safety_stock
reorder_qty    = avg_daily_demand * 30      # 30-day stock

print(f"Avg Daily Demand : £{avg_daily_demand:,.2f}")
print(f"Safety Stock     : £{safety_stock:,.2f}")
print(f"Reorder Point    : £{reorder_point:,.2f}")
print(f"Recommended Order: £{reorder_qty:,.2f}")

results = pd.DataFrame({'Metric': ['Avg Daily Demand','Safety Stock','Reorder Point','Reorder Qty'],
                        'Value': [avg_daily_demand, safety_stock, reorder_point, reorder_qty]})
results.to_csv('../data/inventory_recommendations.csv', index=False)
```
